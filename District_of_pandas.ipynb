```python
#total number of schools
total_schools = school_data["school_name"].count()
print(total_schools)
```

```python
#total number of students
total_students = student_data["Student ID"].count()
print(total_students)
```

```python
#total budget
total_budget = school_data["budget"].sum()
print(total_budget)
```

```python
#total math score
avg_math_score = student_data["math_score"].mean() 

#total reading score
avg_reading_score = student_data["reading_score"].mean()
print(avg_math_score)
print(avg_reading_score)
```

```python
#overall passing rate 
overall_passing_rate = (avg_math_score + avg_reading_score)/2
print(overall_passing_rate)
```

```python
pass_math = student_data.loc[(student_data["math_score"] >= 70)]
total_pass_math = pass_math["math_score"].count()
#print(total_pass_math)
percent_pass_math = (total_pass_math/total_students)*100
print(percent_pass_math)
```

```python
pass_reading = student_data.loc[(student_data["reading_score"] >= 70)]
total_pass_reading = pass_reading["reading_score"].count()
#print(total_pass_reading)
percent_pass_reading = (total_pass_reading/total_students)*100
print(percent_pass_reading)
```

```python
#overall passing math+reading
overall_passing_percentage = (percent_pass_math + percent_pass_reading)/2
print(overall_passing_percentage)
```

```python
#build data frame to display all info
District_summary = {"Total Schools" : total_schools,
                    "Total Students": total_students,
                    "Total Budget": total_budget,
                    "Average Math Score": avg_math_score,
                    "Average Reading Score": avg_reading_score,
                    "% Passing Math": percent_pass_math,
                    "% Passing Reading": percent_pass_reading,
                    "Overall Passing Rate" : overall_passing_percentage}
District_summary_df = pd.DataFrame([District_summary])
District_summary_df = District_summary_df[["Total Schools", 
                                           "Total Students", 
                                           "Total Budget",
                                          "Average Math Score",
                                          "Average Reading Score",
                                          "% Passing Math",
                                          "% Passing Reading",
                                          "Overall Passing Rate"]]
District_summary_df
```

```python
school = school_data.rename(columns={'school_name' : 'School','type' : 'Type','size' : 'Size','budget' : 'Budget'})
#school.head()
```

```python
school_types = school.set_index(["School"])["Type"]
#school_types.head()
```

```python
#renaming all the columns to make it easier to reference. 
complete_school_data = complete_school_data.rename(columns = {'student_name': 'Student', 
                                                             'school_name' : 'School Name',
                                                             'reading_score': 'Reading Score',
                                                             'math_score': 'Math Score',
                                                             'type': 'Type',
                                                             'size': 'Size',
                                                             'budget': 'Budget'})
complete_school_data.head()
```

```python
#Calculate student per school 
student_per_school = complete_school_data["School Name"].value_counts()
```

```python
#Calculate budget for each student per school 
per_school_budget = complete_school_data.groupby(["School Name"]).mean()["Budget"]
per_student_budget = per_school_budget/student_per_school

#Calculate Avg math score & reading score for each school 
avg_math_score = complete_school_data.groupby(["School Name"]).mean()["Math Score"]
avg_read_score = complete_school_data.groupby(["School Name"]).mean()["Reading Score"]
```

```python
#Calculate %of student that pass math and reading at each school 
#Score >= 70 is passing. 
passing_math = complete_school_data[complete_school_data["Math Score"] >= 70].groupby("School Name").count()["Student"]
passing_reading = complete_school_data[complete_school_data["Reading Score"] >= 70].groupby("School Name").count()["Student"]
```

```python
percent_passing_math = passing_math/student_per_school * 100
percent_passing_reading = passing_reading/student_per_school * 100
overall_passing_rate = (percent_passing_math + percent_passing_reading)/2
#overall_passing_rate
```

```python
#Create school summary data frame
School_summary = pd.DataFrame({"School Type": school_types,
                              "Total Student": student_per_school,
                              "Total Budget": per_school_budget,
                              "Per Student Budget": per_student_budget,
                              "Average Math Score": avg_math_score,
                              "Average Reading Score": avg_read_score,
                              "% Passing Math": percent_passing_math,
                              "% Passing Reading": percent_passing_reading,
                              "Overall Passing Rate": overall_passing_rate})
School_summary
```

```python
##Top performing schools
Top_schools = School_summary.sort_values(['Overall Passing Rate'], ascending = False)
Top_schools.head(5)
```

```python
#Worst Performing Schools 
Bottom_schools = School_summary.sort_values(['Overall Passing Rate'], ascending = True)
Bottom_schools.head(5)
```

```python
#Scores by School Spending 
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
#Use 4 reasonable bins to group school spending. Include in the table each of the following: 
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)
```

```python
School_summary
School_summary_budget = School_summary.sort_values(["Per Student Budget"], ascending = False)
School_summary_budget
#Lowest budget = 578
```

```python
Spending_bins = [0, 585, 615, 645, 675]
Spending_range = ["<$585", "$585-615", "$615-645", "$645-675"]

#Categorize school based on spending range per student and add to school summary table. 
School_summary["Spending Range per Student"] = pd.cut(per_student_budget, Spending_bins, labels = Spending_range)

Spending_vs_math_score = School_summary.groupby(["Spending Range per Student"]).mean()["Average Math Score"]
Spending_vs_reading_score = School_summary.groupby(["Spending Range per Student"]).mean()["Average Reading Score"]
Spending_vs_math_score_perc = School_summary.groupby(['Spending Range per Student']).mean()['% Passing Math']
Spending_vs_read_score_perc = School_summary.groupby(['Spending Range per Student']).mean()['% Passing Reading']
Overall_passing_rate = (Spending_vs_math_score + Spending_vs_reading_score)/2

#Create a table 

Spending_vs_score = pd.DataFrame({'Avg Math Score': Spending_vs_math_score,
                                 'Avg Reading Score': Spending_vs_reading_score,
                                 '% Passing Math': Spending_vs_math_score_perc,
                                 '% Passing Reading': Spending_vs_read_score_perc,
                                 'Overall Passing Rate':Overall_passing_rate })
Spending_vs_score
```

```python
##Scores by School Size
size_bins = [0, 1000, 2000, 5000]
size_range = ['Small <1000', 'Medium (1000-2000)', 'Large (2000-5000)']
School_summary['School Size'] = pd.cut(School_summary['Total Student'], 
                                       size_bins, labels= size_range)
#School_summary.head()

avg_math_score = School_summary.groupby(['School Size']).mean()['Average Math Score']
avg_reading_score = School_summary.groupby(['School Size']).mean()['Average Reading Score']
percent_passing_math = School_summary.groupby(['School Size']).mean()['% Passing Math']
percent_passing_reading = School_summary.groupby(['School Size']).mean()['% Passing Reading']
percent_passing_overall = School_summary.groupby(['School Size']).mean()['Overall Passing Rate']

Size_score = pd.DataFrame({'Average Math Score': avg_math_score,
                          'Average Reading Score': avg_reading_score,
                          '% Passing Math': percent_passing_math,
                          '% Passing Reading': percent_passing_reading,
                          'Overall Passing Rate': percent_passing_overall})
Size_score
```
